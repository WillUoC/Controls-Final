```python
import sympy as sym
import matplotlib.pyplot as plt
from sympy.plotting.plot import MatplotlibBackend, Plot, plot3d, plot3d_parametric_line


t = sym.symbols("t")
a, b, c = sym.symbols('a, b, c')

x = sym.Function('x')(t)
y = sym.Function('y')(t)
z = sym.Function('z')(t)

P = sym.Matrix([
    [x],
    [y],
    [z]
])

R = sym.Matrix([
    [a],
    [b],
    [c]
])

vector = P-R

t = vector.dot(sym.diff(P))

display(t)
```

```python
x = sym.symbols('x0:3')
y = sym.symbols('y0:3')
z = sym.symbols('z0:3')

t = sym.symbols('t')

p0 = sym.Matrix([
    [x[0]],
    [y[0]],
    [z[0]]
])

p1 = sym.Matrix([
    [x[1]],
    [y[1]],
    [z[1]]
])

p2 = sym.Matrix([
    [x[2]],
    [y[2]],
    [z[2]]
])

Pr = sym.Matrix([
    [a],
    [b],
    [c]
])

B = (1-t)**2*p0 + 2*(1-t)*t*p1+t**2*p2
B_prime = 2*(1-t)*(p1-p0)+2*t*(p2-p1)

P = B
Vd = P-Pr

Vp = Vd.dot(sym.diff(B, t))

display(Vp.collect(t))

display(sym.solve(sym.Eq(Vp, 0), t))

```

```python
def get_sympy_subplots(plot: Plot):
    backend = MatplotlibBackend(plot)

    backend.process_series()
    backend.fig.tight_layout()
    return backend.fig, backend.ax[0]

p0 = sym.Matrix([
    [0],
    [0],
    [0]
])

p1 = sym.Matrix([
    [1.5],
    [1],
    [1.5]
])

p2 = sym.Matrix([
    [3],
    [2],
    [6]
])

B = (1-t)**2*p0 + 2*(1-t)*t*p1+t**2*p2
B_prime = 2*(1-t)*(p1-p0)+2*t*(p2-p1)

display(B)

p = plot3d_parametric_line((B[0], B[1], B[2], (t, 0, 1)), show=False)

fig, ax = fig, ax = get_sympy_subplots(p)
ax.plot(p0[0], p0[1], p0[2], "ro")
ax.plot(p1[0], p1[1], p1[2], "ro")
ax.plot(p2[0], p2[1], p2[2], "ro")
```
