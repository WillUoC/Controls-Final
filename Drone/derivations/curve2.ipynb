```python
import numpy as np
import matplotlib.pyplot as plt

def generate_leminscate(a, height=5, res=100):
    points = np.zeros((res+1, 3))
    for i in range(res):
        t = (i)/res*2*np.pi
        x, y, z = ((a*np.cos(t))/(1+np.sin(t)**2), (a*np.sin(t)*np.cos(t))/(1+np.sin(t)**2), height)
        points[i, :] = [x, y, z]
    t = 2*np.pi
    x, y, z = ((a*np.cos(t))/(1+np.sin(t)**2), (a*np.sin(t)*np.cos(t))/(1+np.sin(t)**2), height)
    points[res, :] = [x, y, z]
    
    c = a/np.sqrt(2)  # Focal distance
    L = 7.416*c  # Arc-length

    tolerance = L/(2*res)

    return(points, tolerance)

points, tol = generate_leminscate(4)

plt.plot(points[:, 0], points[:, 1])
```
